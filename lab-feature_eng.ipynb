```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for missing values**

```python
spaceship.isnull().values.sum()
```

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship.dropna(inplace = True)
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship['Cabin']= spaceship['Cabin'].str.split('/').str[0]
spaceship
```

- Drop PassengerId and Name

```python
spaceship.drop(columns = ["PassengerId", "Name"], inplace=True)
```

- For non-numerical columns, do dummies.

```python
# Separate numerical columns
numerical_df = spaceship.select_dtypes(include=['number'])

# Separate categorical columns
categorical_df = spaceship.select_dtypes(exclude=['number'])

# Convert categorical columns to dummy variables
categorical_dummies_df = pd.get_dummies(categorical_df).astype(int)

# Combine numerical and dummy variables DataFrames
spaceship_d = pd.concat([numerical_df, categorical_dummies_df], axis=1)
spaceship_d
```

**Perform Train Test Split**

```python
features = spaceship_d.drop(columns=['Transported'])
```

```python
target = spaceship_d["Transported"]
target
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test, y_test)
```
